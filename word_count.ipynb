```python
import subprocess
import string
```

```python
input_file = "/user/student/shakespeare/tragedy/macbeth.txt"
cat = subprocess.Popen(['hadoop', 'fs', '-cat', input_file], stdout=subprocess.PIPE)
```

```python
for line in cat.stdout:
    line = line.decode('utf-8').rstrip()
    if len(line) == 0:
        continue
    tokens = line.translate(str.maketrans('', '', string.punctuation)).split()
    print(tokens)
```
